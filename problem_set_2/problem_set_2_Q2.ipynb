```python
# 2a

# Modify alg2 for Keyed Sorting
def keyed_merge_sort(data, key_index=0):
    if len(data) <= 1:
        return data
    else:   # Divide the list into two halves
        split = len(data) // 2
        left = iter(keyed_merge_sort(data[:split], key_index))
        right = iter(keyed_merge_sort(data[split:], key_index))
        result = []
        left_top = next(left)
        right_top = next(right)
        while True:  # Compare elements based on the key_index
            if left_top[key_index] <= right_top[key_index]:
                result.append(left_top)
                try:
                    left_top = next(left)
                except StopIteration:
                    return result + [right_top] + list(right)
            else:
                result.append(right_top)
                try:
                    right_top = next(right)
                except StopIteration:
                    return result + [left_top] + list(left)

# Example usage and testing
patient_data = [
    (3, "Alice"),
    (1, "Bob"),
    (4, "Charlie"),
    (2, "David"),
    (5, "Eve")
]

sorted_data = keyed_merge_sort(patient_data)
print("Sorted patient data:", sorted_data)

# Test with reverse sorting
sorted_data_reverse = keyed_merge_sort(patient_data, key_index=1)
print("Sorted patient data by name:", sorted_data_reverse)

# Test with empty list
empty_data = []
sorted_empty = keyed_merge_sort(empty_data)
print("Sorted empty data:", sorted_empty)

# Test with single item
single_item = [(1, "Single")]
sorted_single = keyed_merge_sort(single_item)
print("Sorted single item:", sorted_single)
```
